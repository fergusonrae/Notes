```python
# How to import
import matplotlib.pyplot as plt

# Get example data
import pandas as pd
from pydataset import data
housing = data('Housing')
print(housing.info())
```

```python
# Line chart
# Arguments are x-values, y-values
# Can scale on the fly

plt.plot(housing.index, housing['lotsize'])
plt.xscale('log')
plt.show()
```

## Histogram
Good for getting an idea about the ditribution of a dataset.

2 most important arguments are hist(x, bins=10)

Bins are 10 by default.

```python
plt.hist(housing['price'], bins=3)
plt.show()
plt.clf()
```

## Customizing

### Axis ticks
Specify what the axis numbers are and what potential strings you want to replace them with.

plt.yticks(LIST_AXIS_VALUE, LIST_AXIS_NAME)
plt.xticks()

Lists should be the same size.

```python
# Axis ticks example
plt.hist(housing['lotsize'], bins=3)
plt.yticks([1, 10, 100, 500], ["one", "ten", "one hundred", "five hundred"])
plt.show()
plt.clf()
```

### Marker / Grid Attributes

* Size
* Color
* Data marks
* Grid lines

```python
# Example of setting size
# Here, the size of the dot corresponds to how many bedrooms the house has
# Raised to the 3rd power to be more dramatic
plt.scatter(housing['price'], housing['lotsize'], s=(housing['bedrooms']**3))
plt.show()
```

```python
# Gridlines and data values
plt.scatter(housing['price'], housing['lotsize'], s=(housing['bedrooms']**3))
plt.text(29000, 8000, 'Putting text here')
plt.grid(True)
plt.show()
plt.clf()
```
