```python
# Get example data
import pandas as pd
from pydataset import data
housing = data('Housing')
print(housing.info())
```

```python
# Getting a series back vs getting a dataframe back
# Add an extra set of brackets around the column name to get a dataframe back
# This turns the number of columns into a list, rather than just an element
print(type(housing['price']))
print(type(housing[['price']]))
```

```python
# Iterate over a dataframe in a for loop
for lab, row in housing.iterrows():
    print("the row's label: " + lab)
    print("the row's data: " + row)

# Code for loop that adds DRIVEWAY column
for lab, row in housing.iterrows():
    housing.loc[lab, "DRIVEWAY"] = row['driveway'].upper()
```

```python
# APPLY()

housing["DRIVEWAY"] = housing['driveway'].apply(str.upper)
```

## Global

Variables can be definied as global in a method. This means, that any change to them in the method will be reflected in the local level variable if called.

## Nonlocal
Variables can also be defined as nonlocal in a method. When stated as this, it meas that any change to the variable will be reflected in lower nested functions

```python
# Global variable example

x_int = 30

def change_x():
    x_int = 45

change_x()
print(x_int)

def change_x_global():
    global x_int
    x_int = 60

change_x_global()
print(x_int)
```

```python
# Nonlocal variable example

def outer():
    """Prints the value of n"""
    n=1
    
    def inner():
        n = 2
    inner()
    print(n)

outer()

def outer_nonlocal():
    """Prints the value of n"""
    n=1
    
    def inner():
        nonlocal n
        n = 2
    inner()
    print(n)

outer_nonlocal()
```
